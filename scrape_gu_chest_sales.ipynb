```python
# Instantiate Beautiful Soup.
soup = BeautifulSoup(html, features='html.parser')
```

```python
# Find progress bar data.
sales_progress = soup.find_all('gu-hex-progress-bar')
sales_progress
```

```python
# Rare chest percent remaining uncleaned
rare_percent_remaining = sales_progress[0]
rare_percent_remaining
```

```python
# Legendary chest percent remaining uncleaned
legendary_percent_remaining = sales_progress[1]
legendary_percent_remaining
```

```python
# Function to extract sales progress bar data from the html scraped from the Gods Unchained Website.
def clean_progress_bar_data(html):
    
    # Find progress bar data.
    attributes = html.attrs
    bar_width = attributes['style']
    
    # Strip unwanted text
    first_strip = bar_width.strip('--barWdidth:')
    second_strip = first_strip.strip('%;')
    
    # Cast the stripped value as a float.
    percent_remaining = float(second_strip)

    return percent_remaining
```

```python
cleaned_rare_remaining = clean_progress_bar_data(rare_percent_remaining)
cleaned_rare_remaining
```

```python
cleaned_legendary_remaining = clean_progress_bar_data(legendary_percent_remaining)
cleaned_legendary_remaining
```
