```python
# Dependencies.
import requests
import json

import pandas as pd
```

```python
# Make the API call to retrieve card rarity data for the Gods Unchained Genesis set.
rarity = requests.get("https://api.godsunchained.com/v0/rarity")
```

```python
# Check the API response header
rarity.headers
```

```python
# Store the API response data in JSON format.
rarity_json = rarity.json()
rarity_json
```

```python
# Check the number of keys (card protos) in the API response.
len(rarity_json.keys())
```

```python
# Convert the API response data into a Pandas dataframe.
rarity_df = pd.DataFrame.from_dict(rarity_json)
rarity_df
```

```python
# Swap rows/columns in the dataframe.
transposed_df = rarity_df.transpose()
transposed_df
```

```python
# Sort the dataframe by proto id in ascending order.
sorted_df = transposed_df.sort_values(by="proto") 
sorted_df
```
